# PostDoc Events Survey 2017

Here are the results from our recent SurveyMonkey events survey. We had 53 responses which is great!


```r
library(ggplot2)
library(dplyr)
library(tidyr)
library(ggsignif)
library(stringr)
data <- read.delim("SurveyResults.txt")
data$NEventTypes = str_count(data$Type, ",")+1
data2 <- read.delim("SurveyResults_byEventType.txt")

library(repr)
options(repr.plot.width=8, repr.plot.height=5)
```

## Q1. What school division are you in?

```r
ggplot(data, aes(Division, fill=Division))+geom_bar()
ggplot(data, aes(Responder, color=Division))+geom_freqpoly(stat='density')
```

We got a fairly even response from all three school divisions, although 6 people weren't aware which division they are in - these people were also amongst the last to respond. 

## Q2. Were you aware that a PostDoc committee has been created?

```r
ggplot(data, aes(PostdocComm, fill=PostdocComm))+geom_bar()+ scale_fill_discrete(name="",
                         breaks=c("No", "Yes", "YesKnow"),
                         labels=c("No", "Yes", "Yes, and I know who my representative is"))
ggplot(data, aes(PostdocComm, fill=PostdocComm))+geom_bar()+facet_wrap(~Division)+ scale_fill_discrete(name="",
                         breaks=c("No", "Yes", "YesKnow"),
                         labels=c("No", "Yes", "Yes, and I know who my representative is"))
```

Only 3 people don't know we have a postdoc committee, 2 of whom are in IIM and one in COMGEN. Most people know who they're representative is - especially those in PPN.

## Q8. How well do you know other postdocs in the school?

### Thank you to everyone who answered our survey! 

The key points we're going to take from this are:

1. Make sure more people know who their postdoc committee representative is

1. Organise more events for postdocs, such as:
 * Annual/bi-annual research forums
 * Lunchtime and after-hour socials
 * Half-day or shorter specific events that are focused on networking and skill sharing
 * Broader careers events 
3. Get postdocs talking to each other!
