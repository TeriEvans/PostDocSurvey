# PostDoc Events Survey 2017

Here are the results from our recent SurveyMonkey events survey. We had 53 responses which is great!


```r
library(ggplot2)
library(dplyr)
library(tidyr)
library(ggsignif)
library(stringr)
data <- read.delim("SurveyResults.txt")
data$NEventTypes = str_count(data$Type, ",")+1
data2 <- read.delim("SurveyResults_byEventType.txt")

library(repr)
options(repr.plot.width=5, repr.plot.height=3)
```

## Q1. What school division are you in?

```r
ggplot(data, aes(Division, fill=Division))+geom_bar()
ggplot(data, aes(Responder, color=Division))+geom_freqpoly(stat='density')
```

We got a fairly even response from all three school divisions, although 6 people weren't aware which division they are in - these people were also amongst the last to respond.
